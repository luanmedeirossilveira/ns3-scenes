```python
import pandas as pd
```

```python
# Carregar o arquivo CSV
df = pd.read_csv('../../build/examples/scenes/results-scenario1.csv')

print(df)
```

```python
# Gráfico de vazão de pacotes
plt.figure()
plt.plot(df['Packet Size'], df['Throughput (pps)'], label='Com Roteador')
plt.plot(df['Packet Size'], df['Throughput (pps)'], label='Sem Roteador')
plt.plot(df['Packet Size'], df['Throughput (pps)'], label='Teórico')
plt.xlabel('Tamanho do Pacote (bytes)')
plt.ylabel('Vazão de Pacotes (pacotes/s)')
plt.title('Gráfico de Vazão de Pacotes na Rede')
plt.legend()
plt.grid(True)
plt.savefig('throughput_packets.png')
plt.show()
```

```python
# Gráfico de vazão de bytes
plt.figure()
plt.plot(df['Packet Size'], df['Throughput (Bps)'], label='Com Roteador')
plt.plot(df['Packet Size'], df['Throughput (Bps)'], label='Sem Roteador')
plt.plot(df['Packet Size'], df['Throughput (Bps)'], label='Teórico')
plt.xlabel('Tamanho do Pacote (bytes)')
plt.ylabel('Vazão de Bytes (bytes/s)')
plt.title('Gráfico de Vazão de Bytes na Rede')
plt.legend()
plt.grid(True)
plt.savefig('throughput_bytes.png')
plt.show()
```

```python
# Gráfico de utilização da CPU em relação ao tamanho dos pacotes
plt.figure()
plt.plot(df['Packet Size'], df['CPU Utilization (%)'])
plt.xlabel('Tamanho do Pacote (bytes)')
plt.ylabel('Utilização da CPU (%)')
plt.title('Gráfico de Utilização da CPU em Relação ao Tamanho dos Pacotes')
plt.grid(True)
plt.savefig('cpu_utilization.png')
plt.show()
```
